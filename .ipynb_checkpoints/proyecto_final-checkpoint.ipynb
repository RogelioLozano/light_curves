El universo tiene muchos eventos astronómicos con variaciones temporales en su brillo. 

En general se consideran los eventos como transitorios (por ejemplo, las supernovas que exhiben un pico en un cierto intervalo temporal y luego se desvanecen para siempre) y los variables (son periodicos como un sistema binario que bloquean la luz del otro perodicamente).

Las variaciones en el brillo pueden dar información importante sobre el evento, su entorno local y la evolución del universo. Como ejemplo considerar que mediciones sobre curvas de luz de supernovas de tipo Ia sirvieron como primera evidencia de la expansión acelerada del universo.

Entonces resulta ser muy importante clasificar adecuadamente los eventos. Se tienen dos formas de obtener información de objetos astronómicos. La más precisa es usando medidas espectroscópicas donde se mide la densidad de flujo espectral en función de la longitud de onda. Problema: requieren mucho tiempo y no es factible con la cantidad de información que será generada por LSST. Nos queda entonces usar medidas fotométricas en las que se toman fotos a través de distintos filtros con un cierto rango de longitud de onda. 

Factores generales a tomar en cuenta:


+ Los filtros también llamadas pasabandas del LSST son 6, denotados u,g,r,i,z,y que pasan luz en los rangos 300 y 400 nanometros para la banda u, 400 y 600 nm para la banda g, 500 y 700 nm para la r, 650 y 850 nm para la banda i, 800 y 950 nm para la banda z y 950 y 1050 nm para la banda y. La Figura 2 muestra dos tipos de bandas para un evento de supernova tipo Ia. 




+ El corrimiento al rojo disminuye la tasa de llegada de fotones comparada con la tasa a la que fueron emitidos. Debido a que los objetos con un corrimiento al rojo mayor estan más lejanos, la luz que se recibe de ellos es más tenue. Por lo tanto, las observaciones de objetos lejanos que pasan el filtro noise to signal ratio pueden ser menos que las de objetos cercanos. La Figura 2 ilustra el efecto del corrimiento al rojo para un evento de supernova tipo Ia. La linea negra sólida representa el espectro del evento de supernova "cercana" a una distancia de 140 millones de años luz que corresponde a un corrimiento al rojo (z) de 0.01. La linea punteada es el mismo tipo de evento pero a una distancia de 5.1 billones de años luz. Notar que la el espectro se mueve hacia el rojo



+ El LSST tendrá 2 tipos de regiones de reconocimiento. La primera es el Deep Drilling Fields (DDF) que son partes pequeñas del cielo que seran sampledas frecuentemente para obtener mayor profunidad (i.e., para poder medir el flujo de objetos más tenues). Las curvas de luz de estas regiones van a estar muy bien determinadas y tendran poca incertidumbre las mediciones. La segunda es el Wide-Fast-Deep (WFD) que cubrirá mucha mayor área del espacio, casi 400 veces el área del DDF, pero se visualizara con menor frecuencia, por lo tanto aumentando las incertidumbres de las curvas de luz.

```python
train_metadata['ddf'].value_counts(normalize=True)*100
indx = train_metadata[train_metadata['ddf']==1]['object_id']
```

```python
train_df_DDF=train_df[train_df['object_id'].isin(indx)]
```

```python
groups = train_df_DDF.groupby(['object_id', 'passband'])
def normalise(ts):
    return (ts - ts.mean()) / ts.std()

times = groups.apply(
    lambda block: block['mjd'].values).reset_index().rename(columns={0: 'seq'})

flux = groups.apply(
    lambda block: normalise(block['flux']).values
).reset_index().rename(columns={0: 'seq'})

times_list = times.groupby('object_id').apply(lambda x: x['seq'].tolist()).tolist()
flux_list = flux.groupby('object_id').apply(lambda x: x['seq'].tolist()).tolist()
```

```python
def phase_lc(n,freq,ax):
    colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple']

    for band in range(6):
        ax.plot(times_list[n][band] * freq % 1,flux_list[n][band],
                'o',c=colors[band],label=colors[band][0])
    ax.set_xlabel('phase')
    ax.set_ylabel('flux')
    ax.set_title('class:{}'.format(train_metadata['target'][n]))
```

```python
import cesium.featurize as featurize
warnings.simplefilter('ignore', RuntimeWarning)

N=200
cfeats=featurize.featurize_time_series(times=times_list[:N],
                                      values=flux_list[:N],
                                      features_to_use=[
                                          'freq1_freq','freq1_signif','freq1_amplitude1'
                                      ], scheduler=None)
```

```python
cfeats.head(10)
```

```python
fig,ax=plt.subplots(2,2,figsize=(12,6),sharex=False)
phase_lc(0,3.081631,ax=ax[0,0])
phase_lc(3,0.001921,ax=ax[0,1])
phase_lc(150,1.793083,ax=ax[1,1])
phase_lc(8,0.004020,ax=ax[1,0])
ax[1,1].legend(loc=1, 
             bbox_to_anchor=(0.2, -0.3),fancybox=True, shadow=False, ncol=3)
```

```python
t=np.linspace(0,10*np.pi,1000)
t = np.r_[t,np.linspace(1000*np.pi,1010*np.pi,1000)]
x=np.sin(t)
plt.plot(t,x)
plt.show()
t_periodic = t/(2*np.pi) % 1
plt.plot(t_periodic,x,'o')
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

data_preprocessed=cfeats.stack('channel').reset_index()
train, test = train_test_split(np.arange(N), random_state=0)

predictions={}
for i in range(2):
    
    forest_params={'max_depth':range(6,12),
              'max_features':range(4,19)}
    
    to_fit=data_preprocessed[data_preprocessed.channel==i]
    model = RandomForestClassifier(n_estimators=128,n_jobs=-1, max_features='auto',
                                         random_state=0, criterion='gini')

    forest_grid=GridSearchCV(model,forest_params,cv=5,n_jobs=-1,verbose=True)
    forest_grid.fit(to_fit.iloc[train],train_metadata['target'][train])
    predictions[i] = forest_grid.predict(to_fit)
```

```python
predictions=pd.DataFrame(predictions)
```

```python
pb=1
print("Built-in cesium features: training accuracy={:.2%}, test accuracy={:.2%}".format(
          accuracy_score(predictions.iloc[train,pb], train_metadata['target'][train]),
          accuracy_score(predictions.iloc[test,pb], train_metadata["target"][test])))
```

Stochastic process: Loosely, una coleccion de variables aletorias indexadas:

$$Y=(Y_1,Y_2,Y_3,...,Y_n,...)$$

Por ejemplo, pueden ser variables con distribucion bernoulli que toman los valores de 0 y 1 cada una. Tendriamos en este caso un proceso bernoulli. 

Podemos contruir nuevos procesos a partir de otros. Por ejemplo, considerando el proceso bernoulli anterior, podemos generar un random walker con la suma cumulativa de Y's:

$$X=(Y_0,Y_0+Y_1, Y_0+Y_1+Y_2,...)$$

```python
eeg = datasets.fetch_andrzejak()
```

```python
# Group together signal classes normal(Z,O), interictal(N,F), ictal(S)
eeg['classes'] = eeg['classes'].astype("U16")
eeg["classes"][np.logical_or(eeg["classes"]=="Z", eeg["classes"]=="O")] = "Normal"
eeg["classes"][np.logical_or(eeg["classes"]=="N", eeg["classes"]=="F")] = "Interictal"
eeg["classes"][eeg["classes"]=="S"] = "Ictal"
```

```python
fig, ax = plt.subplots(1, len(np.unique(eeg["classes"])), figsize=(20,3), sharey=True)
for label, subplot in zip(np.unique(eeg["classes"]), ax):
    i = np.where(eeg["classes"] == label)[0][0]
    subplot.plot(eeg["times"][i], eeg["measurements"][i])
    subplot.set(xlabel="time (s)", ylabel="signal", title=label)
```

```python
from cesium import featurize
```

```python
sample_times=np.random.choice(range(len(eeg['times'])),size=100,replace=False)
```

```python
times_s = [ eeg['times'][i] for i in sample_times ]
measurements_s = [ eeg['measurements'][i] for i in sample_times ]
```

```python
features_to_use=['amplitude','percent_beyond_1_std','maximum','max_slope',
                 'median','median_absolute_deviation',
                "percent_close_to_median",
                   "minimum",
                   "skew",
                   "std",
                   "weighted_average"]
fset_cesium = featurize.featurize_time_series(times=times_s,
                                             values=measurements_s,
                                             errors=None,
                                             features_to_use=features_to_use)

print(fset_cesium.head())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

train, test = train_test_split(np.arange(100), random_state=0)

model_cesium = RandomForestClassifier(n_estimators=128, max_features='auto',
                                     random_state=0, )
model_cesium.fit(fset_cesium.iloc[train], eeg['classes'][train])
```

```python
from sklearn.metrics import accuracy_score

preds_cesium = model_cesium.predict(fset_cesium)

print("Built-in cesium features: training accuracy={:.2%}, test accuracy={:.2%}".format(
          accuracy_score(preds_cesium[train], eeg["classes"][train]),
          accuracy_score(preds_cesium[test], eeg["classes"][test])))
```
